```python
import numpy as np
import matplotlib.pyplot as plt

from numpy import pi
x = np.linspace(0, 2 * pi, 9)
print(x)
f = np.tan(x)

plt.plot(x, f)
plt.title("Sine Wave")
plt.xlabel("x (radians)")
plt.ylabel("sin(x)")
plt.grid(True)
plt.show()
```
